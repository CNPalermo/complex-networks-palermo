```python
import random

# Define the number of nodes in the core and periphery
num_core_nodes = 10
num_periphery_nodes = 50

# Define the probability of connecting two core nodes
core_edge_prob = 0.8

# Define the probability of connecting a periphery node to a core node
periphery_edge_prob = 0.2

# Create a list of nodes
nodes = list(range(num_core_nodes + num_periphery_nodes))

# Create an empty list to store the edges
edges = []

# Connect the core nodes to each other with the specified probability
for i in range(num_core_nodes):
    for j in range(i+1, num_core_nodes):
        if random.random() < core_edge_prob:
            edges.append((i,j))

# Connect the periphery nodes to the core nodes with the specified probability
for i in range(num_core_nodes, num_core_nodes+num_periphery_nodes):
    for j in range(num_core_nodes):
        if random.random() < periphery_edge_prob:
            edges.append((i,j))

# Shuffle the edges randomly
random.shuffle(edges)

# Print the edges as an edgelist


file=open("core_periphery.txt", "w")


for edge in edges:
    file.write(f"{edge[0]} {edge[1]}\n")

file.close()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.read_edgelist("core_periphery.txt")
nx.draw(G)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Read edge list from file
with open('core_periphery.txt', 'r') as f:
    edgelist = [tuple(map(int, line.strip().split())) for line in f]

# Create empty graph
G = nx.Graph()

# Add edges to graph from edgelist
G.add_edges_from(edgelist)

# Draw graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)

# Show graph
plt.show()
```
